## Line Charts

- Show trends over time
- TIme series data etc

````python
fmri = sns.load_dataset("fmri")
fmri.head()
````

````python
sns.lineplot(x="timepoint", y="signal",data=fmri)
````

````python
sns.lineplot(x="timepoint", y="signal",data=fmri)
# since seaborn is built on mtplotlib we can use plt functions for customisation
plt.title("FRMI Signal Acress Time", fontsize=14, fontweight='bold')
plt.xlabel('Timepoint', fontsize=12)
plt.ylabel('% Signal Change', fontsize=12)
plt.grid(True, alpha=0.3)
````

````python
sns.lineplot(x="timepoint", y="signal", hue="event",
             palette=['teal','indigo'], data=fmri)

plt.title('FMRI Signal by Event Type')
plt.legend(title='Event Type')  # Customize legend title
````

That was all with seaborn, but what if we want to just use raw matplotlib...?

````python
plt.plot(fmri.timepoint, fmri.signal)
plt.title('Matplotlib Line Plot (Raw Data - Notice the Messiness!)\nConnecting all 1,064 points without averaging')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
````

````python
# sales data

new_df = pd.DataFrame({
    'years':[2014,2015,2016,2017,2018,2019,2020],
    'sales':[4800,4900,5000,5700,6500,5000,6700]
})

new_df
````

Use plt ONLY to create a line plot. plt.plot() to show sales over time

````python
plt.plot(new_df.years, new_df.sales)
plt.title("Sales Over Time")
plt.xlabel("Years")
plt.ylabel("Sales ($)")
plt.grid(False, alpha=0.3)
````

````python
plt.plot(new_df.years, new_df.sales, marker='o', linewidth=2,
         markersize=8, color='steelblue')
plt.title("Sales Over Time (Enhanced)", fontsize=14, fontweight='bold')
plt.xlabel("Years", fontsize=12)
plt.ylabel("Sales ($)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
````

- Create a single bar chart showing all 10 consoles with their sales, colored by manufacturer. KISS!

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Top 10 Best-Selling Video Game Consoles Dataset
console_data = {
    'console': [
        'PlayStation 2', 'Nintendo DS', 'Game Boy/Game Boy Color',
        'PlayStation 4', 'PlayStation', 'Nintendo Switch',
        'Wii', 'PlayStation 3', 'Xbox 360', 'Game Boy Advance'
    ],
    'sales_millions': [155, 154.02, 118.69, 117.2, 102.49, 132.46, 101.63, 87.4, 85.8, 81.51],
    'manufacturer': [
        'Sony', 'Nintendo', 'Nintendo', 'Sony', 'Sony',
        'Nintendo', 'Nintendo', 'Sony', 'Microsoft', 'Nintendo'
    ],
    'release_year': [2000, 2004, 1989, 2013, 1994, 2017, 2006, 2006, 2005, 2001],
    'generation': [6, 7, 4, 8, 5, 9, 7, 7, 7, 6],
    'type': [
        'Home Console', 'Handheld', 'Handheld', 'Home Console',
        'Home Console', 'Hybrid', 'Home Console', 'Home Console',
        'Home Console', 'Handheld'
    ],
    'discontinued': [2013, 2013, 2003, 2021, 2006, False, 2013, 2017, 2016, 2010]
}

# Create DataFrame
consoles_df = pd.DataFrame(console_data)

# Sort by sales for better visualization
consoles_sorted = consoles_df.sort_values('sales_millions', ascending=False)

plt.figure(figsize=(14, 7))

sns.barplot(data=consoles_sorted, x='console', y='sales_millions',
            hue='manufacturer', palette='Set1', dodge=False)

plt.title('Top 10 Best-Selling Video Game Consoles', fontsize=16, fontweight='bold')
plt.xlabel('Console', fontsize=13)
plt.ylabel('Sales (Millions)', fontsize=13)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Manufacturer', loc='upper right', fontsize=11)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

## Stacked bar charts with matplotlib


````python
# split the data by gender
male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']

# group by day then calculte means
mday = male.groupby('day').mean(numeric_only=True)
fday = female.groupby('day').mean(numeric_only=True)

# merge the data frames
totals = mday.merge(fday, right_index=True, left_index=True)


# calculate totals and proportions
totals['total'] = totals['tip_x'] + totals['tip_y']
totals['prop_mal'] = np.true_divide(totals['tip_x'], totals['total']) * 100
totals['prop_fem'] = np.true_divide(totals['tip_y'], totals['total']) * 100
totals
````

````python
plt.bar(totals.index, totals.prop_mal, label='Male', color='gold',
        bottom=totals.prop_fem)
plt.bar(totals.index, totals.prop_fem, label='Female', color='green')

plt.title('Tip Proportion by Gender and Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Percentage of Total Tips', fontsize=12)
plt.legend(title='Gender', loc='upper right')
plt.grid(axis='y', alpha=0.3)
````
