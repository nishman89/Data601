## CSV files

`pd.read_csv()` - loads csv into a `DataFrame`

- `sep` (delimeter)
- `header` (which row contains column names (defualt is inferred based on first row)
- `na_values` (treat certais string as missing values

```python
iris = pd.read_csv("iris.csv")
iris.head(10)
```

```python
iris = pd.read_csv("iris.csv", sep=",", header="infer")
iris.head()
```

### Writing to CSV

- `to_csv` to write a dataframe to a file. Parameters:
    - `index` (include index columns or not)
    - `sep` (delimter)

```python
iris.to_csv("new_iris.csv", index=False)
# import os
# os.getcwd()
```

## `groupby`

- Split our dataframe into groups based on a key columns. In our case it will be **species**
- Remember `groupby` does NOT immediately compute something. We need to apply an aggregatation function (`sum` etc)

```python
spec = iris.groupby("species")
spec
```

```python
spec.head(10)
```

```python
spec.mean()
```
