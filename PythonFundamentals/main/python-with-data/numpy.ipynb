## `ndarray`

EVERYTHING in NumPy revolves around this - `n-dimensional array`

```python
list1 = [10,20,30]
list1
```

```python
arr1 = np.array(list1)
arr1
```

List stores REFERENCES to objects, but ndarray store RAW numeric values

```python
arr2 = np.arange(10,20,1) #generated dates - numpy prefers this over manually typed values. zeroes, ones. 3 args: first value, up until 20 (2nd arg), 3rd - how much each value increases by
arr2
```

```python
list2 = [40,50,60]
arr3 = np.array([list1, list2])
arr3
```
