# DataFrames

These are 2-d, size-mutable, mixed, tabular strucutres

Under the good:
- Each column is like series
- Columns can have different dtypes
- there ins shared row index

```python
data = [["Atanaska", 20],["Abinash",12],["Erik",23]]
df = pd.DataFrame(data,columns=["Name","Height"])
df
```

```python
df.index
```

```python
# from a dictionart
new_dict = {
    "Name": ["Alpha", "Beta", "Gamma", "Delta"],
    "Sales":[250,300,350,400]
}
df_dict = pd.DataFrame(new_dict)
df_dict
```

## Add a custom index

The index is the ROW LABEL - this can be ranks, IDs dates, etc

```python
df_dict_index = pd.DataFrame(new_dict, index=["rank1", "rank2","rank3","rank4"])
df_dict
```

## From list of dictionaries

```python
dict_list = [
    {"Name": "Alpha", "Sales":250},
    {"Name": "Beta", "Sales":450},
    {"Name": "Gamma", "Sales":650},
    {"Name": "Delta", "Sales":850}
]

df_dict_list = pd.DataFrame(dict_list,index=["rank1", "rank2","rank3","rank4"])
df_dict_list
```

## DataFrame from a dictionary of series (alignment by index)

```python
east = pd.Series([1000, 1200, 3400], index=["Q1", "Q2", "Q3"])
east
```

```python
west = pd.Series([1100, 1300, 2400, 3500], index=["Q1", "Q2", "Q3", "Q4"])
west
```

```python
df_region = pd.DataFrame({"East": east, "West":west})
df_region
```

## Adding a column

```python
df_region["North"] = [2000,3000,4000,5000]
df_region["South"] = [5435,543543,4760,7000]
df_region
```

## Shifting and changing the index

- Create a column
- set it as the index

```python
years = ["2016","2017","2018","2019"]
df_region["years"] = years
df_region
```

```python
df_region = df_region.set_index("years")
df_region
```
