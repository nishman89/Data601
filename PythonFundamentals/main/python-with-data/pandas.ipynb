## Reindexing

`reindex` that changes the index to a new set of labels:

- Take the new index you ask for
- Aligns existing data by label
- Add rows for missing labels
- Inserts NaN where data does not exist
- Drops and old labels not in de the index

````python
df_region
````

````python
new_df = df_region.reindex(["2017","2018","2019","2020","2021"])
new_df
````

### Reindex columns


- reorder existing
- add new columns (filled by Nan)

````python
re_indexed = new_df.reindex(columns=["North", "East", "South", "New"])
re_indexed
````

## Missing Data

````python
re_indexed = re_indexed.fillna(0)
re_indexed
````

````python
re_indexed = re_indexed.interpolate()
re_indexed
````

````python
# Drop missing data

re_indexed = re_indexed.dropna(axis=1, how='all')
re_indexed
````
