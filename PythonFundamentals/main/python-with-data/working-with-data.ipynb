## Merging DataFrames

You can merge when you have *related* data split across multiple DataFrames. Two common scenarios:

1. Merge on Index - noth DFs line up by their index
2. Merge on key columns -  joins like SQL using IDs

```python
sepal_info = iris[["sepal_length", "sepal_width", "species"]].copy()
petal_info = iris[["petal_length", "petal_width", "species"]].copy()

# Merging on their row index NOT a column
# inner keeps only row where the index exists in both data frames
# left_index - and right_index = true. Tells pandas to join on the index (i.e.g use sepal_info index as the join key, and petal_info index as the join

merge_on_index = pd.merge(sepal_info,petal_info,how="inner", left_index=True, right_index=True)
merge_on_index
```

### Merge on key columns


```python
sepal_info_with_id = sepal_info.copy()
petal_info_with_id = petal_info.copy()

# add a column called sepal_id
sepal_info_with_id["sepal_id"] = np.arange(len(sepal_info_with_id)) # generate values from 0 to number of rows in the df -1
petal_info_with_id["petal_id"] = np.arange(len(petal_info_with_id)) # generate values from 0 to number of rows in the df -1
petal_info_with_id
```

```python
# merge the two data frames

merged_on_ids = pd.merge(sepal_info_with_id,petal_info_with_id,left_on="sepal_id",right_on="petal_id",how="inner")
merged_on_ids
```

## Combining Data Frames

`combine_first` - fill in missing values in one DataFrame with values from another position

```python
sales1 = pd.DataFrame({
    "East":  [1500, np.nan, 3500, np.nan],
    "West":  [np.nan, 4500, np.nan, 5500],
    "South": [6500, np.nan, 7500, np.nan]
})


sales1
```

```python
sales2 = pd.DataFrame({
    "East": [1200, 1300, np.nan, 1600, 1800],
    "West": [np.nan, 5000, 6000, 7000, 8000]
})
sales2
```

```python
combined_sales = sales1.combine_first(sales2)
combined_sales
```

3 different combinations methods: `merge`,`concat` and `combine_first`

## Reshaping with pivot, stack and unstack

- `Long format` - one measurement per row (e.g. year, month, passenger)
- `Wide format` - the values are spread across columns (e.g. months as columns)

We will be using `seaborn` which is used for data viz

It also comes along with a few sample datasets

```python
import seaborn as sns
print(sns.get_data_home()) #where this data lies
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
## PivoT - long to wide
# index=year -> row becomes year
# columns="month" -> each month becomes a column
# values="passengers" -> the cell values are passenger count

flights_piv = flights.pivot(
    index="year",
    columns="month",
    values="passengers"
)

flights_piv
```

```python
# Stack: wide -> long

# stack() moves the columns labels (months) down into the index
# This create a multiindex (year, month)
# the result is a series NOT a dataframe

flight_stack = flights_piv.stack()
flight_stack.head(10)
```

```python
flights_piv.index
```

```python
flight_stack.index
```

```python
# name the unnamed columns
flight_stack = flight_stack.rename("passengers")
flight_stack
```

```python
## Unpivot long->wide

flights_unstack = flight_stack.unstack()

flights_unstack.head()
```
