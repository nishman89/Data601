- Create a single bar chart showing all 10 consoles with their sales, colored by manufacturer. KISS!

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Top 10 Best-Selling Video Game Consoles Dataset
console_data = {
    'console': [
        'PlayStation 2', 'Nintendo DS', 'Game Boy/Game Boy Color',
        'PlayStation 4', 'PlayStation', 'Nintendo Switch',
        'Wii', 'PlayStation 3', 'Xbox 360', 'Game Boy Advance'
    ],
    'sales_millions': [155, 154.02, 118.69, 117.2, 102.49, 132.46, 101.63, 87.4, 85.8, 81.51],
    'manufacturer': [
        'Sony', 'Nintendo', 'Nintendo', 'Sony', 'Sony',
        'Nintendo', 'Nintendo', 'Sony', 'Microsoft', 'Nintendo'
    ],
    'release_year': [2000, 2004, 1989, 2013, 1994, 2017, 2006, 2006, 2005, 2001],
    'generation': [6, 7, 4, 8, 5, 9, 7, 7, 7, 6],
    'type': [
        'Home Console', 'Handheld', 'Handheld', 'Home Console',
        'Home Console', 'Hybrid', 'Home Console', 'Home Console',
        'Home Console', 'Handheld'
    ],
    'discontinued': [2013, 2013, 2003, 2021, 2006, False, 2013, 2017, 2016, 2010]
}

# Create DataFrame
consoles_df = pd.DataFrame(console_data)

# Sort by sales for better visualization
consoles_sorted = consoles_df.sort_values('sales_millions', ascending=False)

plt.figure(figsize=(14, 7))

sns.barplot(data=consoles_sorted, x='console', y='sales_millions',
            hue='manufacturer', palette='Set1', dodge=False)

plt.title('Top 10 Best-Selling Video Game Consoles', fontsize=16, fontweight='bold')
plt.xlabel('Console', fontsize=13)
plt.ylabel('Sales (Millions)', fontsize=13)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Manufacturer', loc='upper right', fontsize=11)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

## Stacked bar charts with matplotlib


````python
# split the data by gender
male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']

# group by day then calculte means
mday = male.groupby('day').mean(numeric_only=True)
fday = female.groupby('day').mean(numeric_only=True)

# merge the data frames
totals = mday.merge(fday, right_index=True, left_index=True)


# calculate totals and proportions
totals['total'] = totals['tip_x'] + totals['tip_y']
totals['prop_mal'] = np.true_divide(totals['tip_x'], totals['total']) * 100
totals['prop_fem'] = np.true_divide(totals['tip_y'], totals['total']) * 100
totals
````

````python
plt.bar(totals.index, totals.prop_mal, label='Male', color='gold',
        bottom=totals.prop_fem)
plt.bar(totals.index, totals.prop_fem, label='Female', color='green')

plt.title('Tip Proportion by Gender and Day', fontsize=14, fontweight='bold')
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Percentage of Total Tips', fontsize=12)
plt.legend(title='Gender', loc='upper right')
plt.grid(axis='y', alpha=0.3)
````

## Histogams

Show the distribution of a single numerica variable by dividing the data into "bins" and counting the observation of each bin

- Understand the distribution of the data
- Identify skewness, ourliers
- Bell curve

````python
data = np.random.randn(100)
data
````

````python
plt.hist(data)
plt.title('Basic Histogram (10 bins)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
````

````python
plt.hist(data, bins=15, color='steelblue', edgecolor='black')
plt.title('Histogram with 15 Bins', fontsize=14, fontweight='bold')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
````

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(data, bins=5, color='coral')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')


axes[1].hist(data, bins=15, color='steelblue')
axes[1].set_title('15 Bins (Good Balance)')
axes[1].set_xlabel('Value')

axes[2].hist(data, bins=50, color='lightgreen')
axes[2].set_title('50 Bins (Too Many)')
axes[2].set_xlabel('Value')
````

#### Distribution plots with seaborn

````python
sns.histplot(data, kde=True, color='purple')

plt.title('Seaborn Distribution Plot', fontsize=14, fontweight='bold')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
````

## Pairplot

Creates a matric of plots showing:

- Scatter plots for each pair of variables
- Histoagams along the diagnol

A pair plot automatically
- Take all numerica columns
- Creates a grid of plots
- Compares each variable against every other

In the iris data set, it comapres:
- Sepal length
- Sepal width
- Petal length
- Petal width

````python
sns.pairplot(
    iris,                  # The full DataFrame
    hue='species',         # Colour the points by species (so we can see group differences)
    palette='Set2',        # Choose a colour scheme
    height=2.5             # Size of each small subplot
)

plt.suptitle('Iris Dataset - Complete Pairwise Relationships', y=1.02)
````

## Heat maps



````python
flights = sns.load_dataset("flights")
flights
````

````python
flights_pivot = flights.pivot(
    index='year', # an index is the row labels of a dataframe. by default, it is 0, 1, 2, 3 etc
    columns='month', # The column that will become the column headers. Each unieuq month will be a column header
    values='passengers'
)
flights_pivot
````

````python
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, cmap='YlOrRd')
plt.title('Flight Passengers by Year and Month', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.tight_layout()
````

````python
# enhanced heat map


plt.figure(figsize=(14, 10))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='summer',
           linewidths=0.5, linecolor='gray',
           cbar_kws={'label': 'Number of Passengers'})

# plt.title('Monthly Airline Passengers (1949-1960)', fontsize=16, fontweight='bold')
# plt.xlabel('Month', fontsize=12)
# plt.ylabel('Year', fontsize=12)
# plt.tight_layout()
````

## Cluster maps



````python
# SciPy is required for hierarchical clustering
# Seaborn's clustermap uses SciPy to calculate distances between rows/columns
import scipy
import seaborn as sns
import matplotlib.pyplot as plt



# Create a clustermap (a heatmap with hierarchical clustering)
# We are clustering YEARS only, not months

sns.clustermap(
    flights_pivot,          # Wide-format matrix (rows = years, columns = months)

    col_cluster=False,      # Do NOT cluster columns
                            # This keeps months in chronological order (Jan → Dec)

    cmap='RdYlBu_r',        # Colour map:
                            # Red = higher passenger numbers
                            # Blue = lower passenger numbers
                            # "_r" reverses the colour scale

    figsize=(12, 10)        # Size of the full figure (including dendrogram)
)

# Add a title above the entire figure
plt.suptitle(
    'Clustered by Year Only - Months in Order',
    y=0.98                   # Move title slightly down so it doesn't overlap
)
````
